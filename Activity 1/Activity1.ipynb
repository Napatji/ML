Clean data

```python
sample = pd.read_csv("Data_example.csv")
sample.X = sample.X.fillna(0)
new_X = []
for i in sample.X:
    if type(i) != int and str(i).isnumeric() == True:
        i = int(i)
        new_X.append(i)
    else:
        i = 0
        new_X.append(i)
sample.X = new_X
sample.Y = sample.Y.fillna(0)
new_Y = []
for i in sample.Y:
    if type(i) != float and str(i).isnumeric() == True:
        i = float(i)
        new_Y.append(i)
    else:
        i = float(0)
        new_Y.append(i)
sample.Y = new_Y
sample.Z = sample.Z.str.encode('ascii','ignore').str.decode('ascii')
sample
```
